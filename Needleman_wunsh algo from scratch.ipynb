```python
#Needleman Wunch algorithm
from pprint import *

seq1="ATCGT"
seq2="ACGT"

print("The length of seq1 and seq2 are:")
m=len(seq1)
n=len(seq2)

# scoring scheme
match_reward=1
mismatch=-1
gap=-2

print(m,n)
mat=[[0 for i in range(len(seq2)+1)]for j in range (len(seq1)+1)]
pprint(mat)

# fill the 0th row and 0th col
for i in range(1,len(seq1)+1):
    mat[i][0]=i*gap
for i in range(1,len(seq2)+1):
    mat[0][i]=i*gap
print("The matrix after filling the oth row and oth col")
pprint(mat)

#fill the matrix
for i in range(1, len(seq1)+1):
    for j in range(1, len(seq2)+1):
        if(seq1[i-1]==seq2[j-1]):
            mat[i][j]=max(mat[i-1][j-1]+match_reward, mat[i-1][j]+gap, mat[i][j-1]+gap)
        else:
            mat[i][j]=max(mat[i-1][j-1]+mismatch, mat[i-1][j]+gap, mat[i][j-1]+gap)
print("The matrix after filling:")
pprint(mat)

#Now trace backe

string1=""
string2=""

i=len(seq1)
j=len(seq2)

while(i>0 and j>0):
    if(seq1[i-1]==seq2[j-1]):
        if(mat[i][j]==mat[i-1][j-1]+match_reward):
            string1=seq1[i-1]+string1
            string2=seq2[j-1]+string2
            i-=1
            j-=1
            
        elif(mat[i][j]==mat[i][j-1]+gap):
            string1="-"+string1
            string2=seq2[j-1]+string2
            j=j-1
            
        else:
            string1=seq1[i-1]+string1
            string2="-"+string2
            i-=1
    else:
        if(mat[i][j]==mat[i-1][j-1]+mismatch):
            string1=seq1[i-1]+string1
            string2=seq2[j-1]+string2
            i-=1
            j-=1
            
        elif(mat[i][j]==mat[i][j-1]+gap):
            string1="-"+string1
            string2=seq2[j-1]+string2
            j=j-1
            
        else:
            string1=seq1[i-1]+string1
            string2="-"+string2
            i-=1
            
print("The Final string alignment is:")
print(string1)
print(string2)
            

    
```
