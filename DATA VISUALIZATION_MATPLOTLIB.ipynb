# Histogram
plt.hist()
creating dummy data using normal distribution

```python
import numpy as np
x_snd=np.random.randn(500) #genrate data points in standard normal distribution mea=0,sd=1
x_snd
```

```python
math_mu=70
math_sig=8
math_marks=np.round(x_snd*math_sig+math_mu)# we will keep thee marks in between 8 to 70  
# math_marks
phy_mu=50
phy_sig=10
phy_mark=np.round(x_snd*phy_sig+phy_mu)
phy_mark
```

```python
plt.hist(math_marks,bins=10)
# plt.hist(phy_mark,bins=10)#this will overlap-use alpha prameter gives seethroughness
plt.hist(phy_mark,bins=10,alpha=0.8)
plt.show
```

# subplot
Create multiple plots in just a single figure

```python
x=np.arange(0,10,0.1)
y1=x**2
y2=x**3
y3=np.sin(x)
y4=np.cos(x)
```

```python
plt.subplot(2,2,1) #here it is rows,col,index like 3col,3row,and index of each plot
plt.plot(x,y1) 

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2, 3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.show() 
```

```python
plt.hist(iris['PetalLengthCm'],bins=20)
plt.xlabel('petal length')
plt.ylabel('Frequency')
plt.show()
# all are following standard noram ldistribution
```

```python
def preprocess(ele):
    return len(ele.strip())
```

```python
movie['movielen']=movie['movie_title'].apply(preprocess) #hence movies have the space in the name..so preprocess
```

```python
#frequency count on y axis-number of movies having x character in their title means movie title with same lenght
plt.hist(movielen,bins=50)
plt.xlabel('movie lenght')
plt.ylabel('frequency')
plt.xticks(range(0,100,5))
plt.show()
```

```python
# scatter way
freq_title={}

for l in movie['movielen']:
    if freq_title.get(l) is None:
        freq_title[l]=1
    else:
        freq_title[l]+=1
```
