```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
#plot distribution of data like histogram
sns.histplot(iris['petal_length'])
plt.show()
```

```python
sns.scatterplot(x=iris['petal_length'],y=iris['petal_width'])
plt.show()
```

```python
sns.scatterplot(x='petal_length',y='petal_width',data=iris,hue='species')
plt.show()
```

```python
# sns.jointplot(x='petal_length',y='sepal_width',data=iris,kind='hex')
sns.jointplot(x='petal_length',y='sepal_width',data=iris,kind='reg')

plt.show()
```

```python
#pair plot
sns.pairplot(data=iris,hue='species')
plt.show
#diagonal have dist plot
```

```python
#categorical plot
# count plot
```

```python
sns.countplot(x='species',data=iris)
plt.show()
```

```python
sns.boxplot(x='species',y='petal_length',data=iris)
plt.show()
#medal line show the mean
# outer lines how the min and the max length (range)
# dimond show the outliers
```

```python
#conclusion petal length is less than 3 then it is setosa
```

```python
sns.violinplot(x='species',y='petal_length',data=iris)
plt.show()
```
