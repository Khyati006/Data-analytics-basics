```python
## merge two data frame
```

```python
df1=pd.DataFrame({'sname':['mohit','jatin','kjf','mohit'],
                 'cgpa':[2,4,5,3]})
df1
```

```python
df2=pd.DataFrame({'tname':['mohit','jatin','kjf','mohit'],
                 'cgpa':[3,6,8,9]})
df2
```

```python
df1.merge(df2,how='inner')#look for common value 
```

```python
df1.merge(df2, how='left') #as cgpa is diff hene nan
```

```python
df1.merge(df2,how='right')
```

```python
df1.merge(df2,how='outer')#combination of both left and right
```
